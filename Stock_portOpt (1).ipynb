```python
import pandas
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()
```

```python
df = pdr.get_data_yahoo(["BRITANNIA.NS","TITAN.NS","ICICIBANK.NS","DRREDDY.NS","EICHERMOT.NS"], start="2000-01-01", end="2019-12-04")
df.head()
```

```python
df1 = df['Close']
df1.head()
```

```python
!pip install PyPortfolioOpt
```

```python
from pypfopt.efficient_frontier import EfficientFrontier 
from pypfopt import risk_models 
from pypfopt import expected_returns 
```

```python
mu = expected_returns.mean_historical_return(df1) 
S = risk_models.sample_cov(df1)
```

```python
S
```

```python
ef = EfficientFrontier(mu, S)
weights = ef.max_sharpe()
cleaned_weights = ef.clean_weights()
print(cleaned_weights)
ef.portfolio_performance(verbose = True) 
```

```python
total_portfolio_value = float(input("Enter annual investment amount: "))
```

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices 

latest_prices = get_latest_prices(df1)

da = DiscreteAllocation(weights, latest_prices, total_portfolio_value )

allocation, leftover = da.lp_portfolio()
print('Discrete allocation: ', allocation)
print('Funds remaining: Rs {:.2f}'.format(leftover))
```
